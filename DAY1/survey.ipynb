```python
# Genderの名寄せ
col = "Gender"
df[col].replace("All","all", inplace=True)
df_gender = pd.crosstab(index=df["treatment"], columns=df[col], margins=True)
display(df_gender.T.sort_values("All",ascending=False))
```
